```python
def create_file(name_file, fat_table, fat_column, dm_table, dm_column,new_columns):
    lista = fat_table[fat_column].unique()
    dimension = dm_table[dm_column]
    df = pd.concat([pd.DataFrame(lista), dimension], ignore_index=True).drop_duplicates().reset_index()
    df.columns = new_columns
    df.to_csv(f'./new_dms/{name_file}.csv', index=False, sep=";")

create_file('dm_instituicao_de_ensino',fat_idd,'id_codigo_intituicao',dm_instituicao_de_ensino,'nome_da_ies',['id_codigo_intituicao','nome_da_ies'])
create_file('dm_instituicao_de_ensino',fat_prouni,'id_codigo_intituicao',dm_instituicao_de_ensino,'nome_da_ies',['id_codigo_intituicao','nome_da_ies'])
create_file('dm_municipio',fat_fies,'id_municipio',dm_municipio,'municipio',['id_municipio','municipio'])
create_file('dm_municipio',fat_prouni,'id_municipio',dm_municipio,'municipio',['id_municipio','municipio'])
create_file('dm_modalidade_de_ensino',fat_prouni,'modalidade_ensino_bolsa',dm_modalidade_de_ensino,'modalidade_de_ensino',['id_modalidade_de_ensino','modalidade_de_ensino'])
create_file('dm_turno',fat_fies,'id_turno',dm_turno,'turno',['id_turno','turno'])
```
